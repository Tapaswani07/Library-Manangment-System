# My Personal Library Manager: A Console App to Organize and Track Your Books Using Python

---

---

```python
# Step 1: Import os module
import os
```

```python
# Step 2: Create empty library list
library = []
```

```python
# Step 3: Function to add books
def add_book(title, author, genre, year, status="unread"):
    book = {
        "title": title.strip(),
        "author": author.strip(),
        "genre": genre.strip(),
        "year": int(year),
        "status": status.strip().lower()
    }
    library.append(book)
    print(f'Book "{book["title"]}" by {book["author"]} added successfully!')
```

```python
# Step 4: Function to list books
def list_books(sort_by_year=False):
    if not library:
        print("Library is empty.")
        return
    books = sorted(library, key=lambda x: x["year"]) if sort_by_year else library
    for idx, book in enumerate(books, 1):
        print(f'{idx}. {book["title"]} by {book["author"]} [{book["genre"]}, {book["year"]}] – {book["status"].capitalize()}')
```

```python
# Step 5: Search by author
def search_by_author(author_name):
    results = [book for book in library if author_name.lower() in book["author"].lower()]
    return results
```

```python
# Step 6: Delete book by title
def delete_book(title):
    global library
    library = [book for book in library if book["title"].lower() != title.lower()]
    print(f'Book "{title}" has been removed (if it existed).')
```

```python
# Step 7: Reading summary
def reading_summary():
    if not library:
        print("Library is empty.")
        return
    read_books = sum(1 for book in library if book["status"] == "read")
    unread_books = len(library) - read_books
    read_percent = (read_books / len(library)) * 100
    unread_percent = (unread_books / len(library)) * 100
    print(f"Read: {read_books} ({read_percent:.2f}%)")
    print(f"Unread: {unread_books} ({unread_percent:.2f}%)")
```

```python
# Step 8: Show uppercase titles
def uppercase_titles():
    titles = [book["title"].upper() for book in library]
    print("Book Titles (UPPERCASE):")
    for t in titles:
        print(f"- {t}")
```

```python
# Step 9: Save library to file
def save_to_file(filename):
    try:
        with open(filename, "w") as f:
            for book in library:
                line = f'{book["title"]},{book["author"]},{book["genre"]},{book["year"]},{book["status"]}\n'
                f.write(line)
        print(f"Saved {len(library)} books to {filename}")
    except Exception as e:
        print(f"Error saving file: {e}")
```

```python
# Step 10: Load library from file
def load_from_file(filename):
    if not os.path.exists(filename):
        print("File not found. Starting with an empty library.")
        return
    try:
        with open(filename, "r") as f:
            for line in f:
                title, author, genre, year, status = line.strip().split(",")
                add_book(title, author, genre, int(year), status)
        print(f"Library loaded successfully from {filename}")
    except Exception as e:
        print(f"There was a problem reading the file: {e}")
```

```python
# Step 11: Menu function
def menu():
    while True:
        print("\n--- Library Menu ---")
        print("1. Add Book")
        print("2. List All Books")
        print("3. List Books Sorted by Year")
        print("4. Search by Author")
        print("5. Delete Book by Title")
        print("6. Show Reading Summary")
        print("7. Show UPPERCASE Titles")
        print("8. Save Library")
        print("9. Load Library")
        print("10. Exit")

        choice = input("Enter your choice (1-10): ")

        if choice == "1":
            title = input("Enter title: ")
            author = input("Enter author: ")
            genre = input("Enter genre: ")
            year = input("Enter year: ")
            status = input("Enter status (read/unread, optional): ") or "unread"
            add_book(title, author, genre, year, status)
        elif choice == "2":
            list_books()
        elif choice == "3":
            list_books(sort_by_year=True)
        elif choice == "4":
            author = input("Enter author name to search: ")
            results = search_by_author(author)
            if results:
                for book in results:
                    print(f'{book["title"]} by {book["author"]}')
            else:
                print("No books found by that author.")
        elif choice == "5":
            title = input("Enter book title to delete: ")
            delete_book(title)
        elif choice == "6":
            reading_summary()
        elif choice == "7":
            uppercase_titles()
        elif choice == "8":
            filename = input("Enter filename to save: ")
            save_to_file(filename)
        elif choice == "9":
            filename = input("Enter filename to load: ")
            load_from_file(filename)
        elif choice == "10":
            print("Goodbye!")
            break
        else:
            print("Invalid option. Try again.")
```

```python
# Preload 5 sample books
add_book("The Great Gatsby", "F. Scott Fitzgerald", "Fiction", 1925, "read")
add_book("1984", "George Orwell", "Dystopian", 1949, "read")
add_book("To Kill a Mockingbird", "Harper Lee", "Fiction", 1960, "unread")
add_book("A Brief History of Time", "Stephen Hawking", "Science", 1988, "unread")
add_book("The Alchemist", "Paulo Coelho", "Fiction", 1988, "read")
```

```python
# Step 12: Run the program
if __name__ == "__main__":
    menu()
```
